```python
from sklearn import datasets
import numpy as np
import tensorflow as tf
import keras
```

```python
data = datasets.load_breast_cancer()
```

```python
print(data.feature_names)
print(data.target_names)
```

```python
featuresData = data.data
targetData = data.target

print(featuresData.shape, targetData.shape)
```

#### menggunakan softmax pada output layers

```python
modelSoft = keras.models.Sequential()
modelSoft.add(keras.layers.Dense(20, activation="relu", input_shape=(30,)))
modelSoft.add(keras.layers.Dense(1, activation="softmax"))
modelSoft.summary()
```

```python
modelSoft.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
modelSoft.fit(featuresData, targetData, batch_size=100, epochs=10)
```

#### menggunakan sigmoid

```python
modelSigm = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(30,)),
    tf.keras.layers.Dense(20, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid"),
])
modelSigm.summary()
```

```python
modelSigm.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
modelSigm.fit(featuresData, targetData, batch_size=100, epochs=10)
```

##### perbedaan softmax dan sigmoid pada output layers

```python
predSoft = modelSoft.predict(featuresData)
predSoft
```

```python
predSigm = modelSigm.predict(featuresData)
predSigm
```

```python
threshold = 0.5
result = (predSigm>threshold).astype(int)
result
```
